## HAPUS STOPWORDS

Penulis menggunakan bantuan library python `stopword NLTK` untuk dapat menggunakan *custom dictionary Indonesia stopword collection `(stopwords-id)`* oleh Gene Diaz yang dapat diakses melalui https://github.com/stopwords-iso/stopwords-id.  Penulis menggunakan dictionary stopword-id dikarenakan kamus tersebut sudah memiliki `MIT License`.

```python
base = output

# Buka file input file dan membacanya 'kata per kata' di tiap baris
input_stream = open(base, 'r')
lines = input_stream.read().splitlines()
input_stream.close()

# Pastikan untuk mengganti path dengan absolute path direktorimu jika baris berikut dijalankan ulang, atau restart kernel.
with open('../Dictionary/StopWord/id.stopwords.02.01.2016.txt') as f:
    custom_stop = f.read().splitlines()
```

```python
# memanfaatkan modul stopwords NLTK untuk menggunakan stopwords kustom
stopwords = stopwords.words('indonesian')
stopwords.clear()
stopwords.extend(custom_stop)
print(stopwords[:3])
```

```python
# Buat file output
output = os.path.splitext(base)[0]+'-stopWord.txt'


with open(output, 'w') as f:
    # Menghapus stopword yang ditemukan di setiap baris
    for line in lines:
        s = ""
        words = line.split()
        for w in words:
            if not w in stopwords:
                s+=str(w)+" "
        res = s

        # Tulis setiap baris yang sudah dikoreksi ke file output
        f.write(str(res+'\n'))
```

```python
# Buat file output
output = os.path.splitext(base)[0]+'-stopWordSastrawati.txt'

factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()

with open(output, 'w') as f:
    # Menghapus stopword yang ditemukan di setiap baris
    for line in lines:
        res= stopword.remove(line)
        
        # Tulis setiap baris yang sudah dikoreksi ke file output
        f.write(str(res+'\n'))
```
